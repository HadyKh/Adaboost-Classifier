```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.ensemble import AdaBoostClassifier
from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split
```

```python
rs = 0
x_circles, y_circles = make_circles(300, noise = 0.1, random_state = rs)
circle_x_train, circle_x_test, circle_y_train, circle_y_test = train_test_split(x_circles, y_circles,
                                                                                test_size = 0.2, random_state = rs)
```

```python
def plotting(x, y, model, title):
    cls =len(set(y))
   
    X, Y = np.array(x), np.array(y) #convert to numpy
   
    plt.figure(figsize=(7,5), dpi=250)
   
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.2),
                         np.arange(y_min, y_max, 0.2))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    cs = plt.contourf(xx, yy, Z, cmap=plt.cm.YlGnBu)

    plt.xlabel('X1')
    plt.ylabel('X2')
    
    plot_colors = ['orange','green','purple']
    # Plot the training points
    for i, color in zip(range(cls), plot_colors):
        idx = np.where(y == i)
        plt.scatter(X[idx, 0], X[idx, 1], c=color,
                    cmap=plt.cm.YlGnBu, edgecolor='black', s=10, alpha=0.6)

    plt.title(title)
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    plt.axis("tight")
    plt.show()
```

```python
def model_Boosting(nEst, lr, x_train, y_train, x_test, y_test):
    estimator = AdaBoostClassifier(n_estimators = nEst, learning_rate = lr, random_state=0)
    score = estimator.fit(x_train, y_train).score(x_test, y_test)
    plotting(x_test, y_test, estimator, f'AdaBoosting: ne={nEst} , lr={lr} , accuracy={score}')

    return estimator
```

```python
ne = [10, 50, 100, 200]
lr = [0.1,1,2]

for i in ne:
    for j in lr:
        model_Boosting(i, j, circle_x_train, circle_y_train, circle_x_test, circle_y_test)
```
